```python
import pandas as pd
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```python
kw_list = ['baseball', 'soccer', 'hockey', 'basketball', 'football']
pytrend.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
```

```python
df=pytrend.interest_over_time()
```

```python
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = df.drop(labels=['isPartial'],axis='columns')
data.to_csv('GoogleTrends_sports.csv', encoding='utf_8_sig')
image = data.plot(title = 'Sports Keyword Interest Over 5 Years on Google Trends ')
fig = image.get_figure()
fig.set_size_inches(18.5, 10.5)
fig.savefig('GoogleTrends_sports.png')
```

```python
#Trending searches
trending_searches = pytrend.trending_searches(pn='united_states')
print(trending_searches)
```

```python
#Top charts
top_charts = pytrend.top_charts(2020, hl='en-US', tz=300, geo='GLOBAL')
print(top_charts)
```
